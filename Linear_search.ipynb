```python
import time
import matplotlib.pyplot as plt

# Linear Search Implementation
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# Measure time complexity
def measure_time_complexity():
    input_sizes = list(range(1, 101))  # Sizes from 1 to 100
    times = []

    # Run linear search on different input sizes
    for size in input_sizes:
        arr = list(range(size))  # Create an array from 0 to size-1
        target = size  # Set the target as the last element in the array

        start_time = time.time()
        linear_search(arr, target)
        end_time = time.time()

        # Record the time taken for each input size
        times.append(end_time - start_time)

    # Plot the graph
    plt.plot(input_sizes, times, label="Linear Search")
    plt.xlabel('Input Size (n)')
    plt.ylabel('Time Taken (seconds)')
    plt.title('Time Complexity of Linear Search')
    plt.legend()
    plt.show()

# Call the function to plot the graph
measure_time_complexity()
```
